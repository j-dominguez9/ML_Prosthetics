```python
# Creating X and y
y = clean_data['act'].values
X = clean_data.drop(columns = ['act']).values
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 21)
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

VW

```python
def to_vw_format(row):
    res = f"{int(row.y)} |"
    for idx, value in row.drop(["y"]).iteritems():
        feature_name = idx
        res += f" {feature_name}:{value}"
    return res
```

```python
import vowpalwabbit

train_vw = X_train_scaled.copy()
test_vw = X_test_scaled.copy()

train_vw = pd.DataFrame(train_vw)
test_vw = pd.DataFrame(test_vw)

train_vw['y'] = y_train
test_vw['y'] = y_test

train_vw_format = train_vw.apply(to_vw_format, axis=1)
test_vw_format = test_vw.apply(to_vw_format, axis=1)

print(train_vw_format)
print(test_vw_format)
```

```python
vw = vowpalwabbit.Workspace('--passes 1000 --cache --oaa 5 --quiet -l 10 --l2 0 -q :: --cubic ::: --loss_function hinge --random_seed 10')

# learn from training set with multiple passes
for sample in train_vw.apply(to_vw_format, axis=1):
    vw.learn(sample)

# predict from the testing set
predictions = []
for sample in test_vw.apply(to_vw_format, axis=1):
    predicted_class = vw.predict(sample)
    predictions.append(predicted_class)

accuracy = len(test_vw[test_vw.y == predictions]) / len(test_vw)
print(f"Model accuracy {accuracy}")
```

```python
from sklearn.metrics import classification_report

true = test_vw['y']
pd.DataFrame(classification_report(true, predictions, output_dict=True))
```
