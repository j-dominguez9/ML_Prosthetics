```python
# Finding outliers
upper_outliers = (data > 50000).any(axis=1)
print(upper_outliers)

lower_outliers = (data < -50000).any(axis=1)
print(lower_outliers)
```

```python
# Filtering out outlier rows
clean_data = data.copy()

clean_data['upper outliers'] = upper_outliers
clean_data['lower outliers'] = lower_outliers
clean_data = clean_data[clean_data['upper outliers'] == False]
clean_data = clean_data[clean_data['lower outliers'] == False]

clean_data = clean_data.drop(columns = ['upper outliers', 'lower outliers'])
clean_data
```

```python
# Creating X and y
y = clean_data['act'].values
X = clean_data.drop(columns = ['act']).values
```

```python
clean_data.shape
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 21, stratify = y)
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

RF Classifier

```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
#from sklearn.externals import joblib
```

```python
clf = RandomForestClassifier(n_estimators=256, criterion='gini', random_state=9, n_jobs = -)
clf.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

NN

```python
import tensorflow as tf
import numpy as np
```

```python
my_model = tf.keras.Sequential()
```

```python
X_train_scaled.shape
```

```python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
```

```python
y_train_cat.shape
```

```python
my_model.add(tf.keras.Input(shape=(39,)))
my_model.add(tf.keras.layers.Dense(400, activation='sigmoid'))
my_model.add(tf.keras.layers.Dense(400, activation = 'sigmoid'))
my_model.add(tf.keras.layers.Dense(500, activation = 'sigmoid'))
my_model.add(tf.keras.layers.Dense(13, activation = 'softmax'))
# my_model.add(tf.keras.layers.Dropout(.1, ))
```

```python
# adam = tf.keras.optimizers.Adam(learning_rate = 0.01, use_ema = True)
adam = tf.keras.optimizers.Adam()
```

```python
my_model.compile(loss='categorical_crossentropy', optimizer= 'adam', metrics=['accuracy', tf.keras.metrics.Precision(class_id =1), tf.keras.metrics.Recall(class_id=1), tf.keras.metrics.Precision(class_id=2), tf.keras.metrics.Recall(class_id=2), tf.keras.metrics.Precision(class_id=3), tf.keras.metrics.Recall(class_id = 3), tf.keras.metrics.Precision(class_id=4), tf.keras.metrics.Recall(class_id=4), tf.keras.metrics.Precision(class_id=5), tf.keras.metrics.Recall(class_id=5), tf.keras.metrics.Precision(class_id=6), tf.keras.metrics.Recall(class_id=6), tf.keras.metrics.Precision(class_id=7), tf.keras.metrics.Recall(class_id=7), tf.keras.metrics.Precision(class_id=8), tf.keras.metrics.Recall(class_id=8), tf.keras.metrics.Precision(class_id=9), tf.keras.metrics.Recall(class_id=9), tf.keras.metrics.Precision(class_id=10), tf.keras.metrics.Recall(class_id=10), tf.keras.metrics.Precision(class_id=11), tf.keras.metrics.Recall(class_id=11), tf.keras.metrics.Precision(class_id=12), tf.keras.metrics.Recall(class_id=12)])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
safety  = EarlyStopping(monitor = 'val_accuracy', patience = 4, mode = 'max')
```

```python
my_model.fit(X_train_scaled, y_train_cat, epochs = 1000, batch_size = 600, validation_data=(X_test_scaled, y_test_cat), callbacks = [safety])
```
